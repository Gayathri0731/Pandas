```python
country_resp = df['Country'].value_counts()
country_resp
```

```python
country_use_python = country_grp['LanguageHaveWorkedWith'].apply(lambda x : x.str.contains('Python').sum())
country_use_python
```

```python
pandas_df = pd.concat([country_resp , country_use_python], axis = 'columns', sort=False)
pandas_df
```

```python
pandas_df.rename(columns = {'count':'No.of.Resp', 'LanguageHaveWorkedWith' : 'Known_Python'},inplace=True)
pandas_df
```

```python
pandas_df['pct_know_python'] = (pandas_df['Known_Python']/pandas_df['No.of.Resp'])*100
pandas_df
```

```python
pandas_df.sort_values(by='pct_know_python',ascending=False)
```
